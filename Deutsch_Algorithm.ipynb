```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
qcir= QuantumCircuit(2,1)
qcir.x(1)
qcir.h([0,1])
qcir.barrier()
qcir.draw(output= 'mpl')
```

```python
qcir.cx([0],[1])
qcir.barrier()
qcir.draw(output= 'mpl')
```

```python
qcir.h([0,1])
qcir.barrier()
qcir.measure(0,0)
qcir.draw(output= 'mpl')
```

```python
simulator= Aer.get_backend('qasm_simulator')
result= execute(qcir,backend= simulator,shots= 1).result()
counts= result.get_counts()
print(counts)
plot_histogram(counts)
```
