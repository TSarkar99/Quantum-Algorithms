```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secretno= '100110100'
n= len(secretno)
```

```python
qc= QuantumCircuit(n+1,n)
qc.x(n)
qc.h(range(n+1))
qc.barrier()
qc.draw(output= 'mpl')
```

```python
for i in range(0,n):
    if secretno[n-i-1]== '1':
        qc.cx([i],[n])
    else:
        pass

qc.barrier()
qc.draw(output= 'mpl')
```

```python
qc.h(range(n))
qc.barrier()
qc.measure(range(n),range(n))
qc.draw(output= 'mpl')
```

```python
simulator= Aer.get_backend('qasm_simulator')
result= execute(qc,backend= simulator,shots= 1).result()
print(result.get_counts(qc))
plot_histogram(result.get_counts(qc))
```
